```python
for i in range(0,len(pokemon_list)):
    path = pokemon_path+pokemon_list[i]
    print("Testing: "+pokemon_list[i])

    for j in os.listdir(path):
        image = cv2.imread(os.path.join(path, j))
            
        try:
            smaller = cv2.resize(image, (96, 96)) 
            X.append(smaller)
        except:
            print(os.path.join(path, j), "Error")
            continue       

    X = np.array(X).reshape(-1, 96, 96, 3)
    X = X / 255.0
    test_results[pokemon_list[i]] = model.predict(X)
```

```python
predicted = {}
evolved = test_results.keys()
count = []

for i in evolved:
    for j in range(0,len(test_results[i])):
        count.append(types[np.argmax(test_results[i][j])])
    predicted[i]=count
    count=[]
```
